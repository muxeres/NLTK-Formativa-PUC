#### 1) Utilize `Expressões Regulares` para remover de cada um dos textos do corpus o cabeçalho dos capítulos (e.g., "*Capítulo 1:*", "*Capítulo 2*:", "*Capítulo 3*:")

```
import re
#identificação dos cabeçalhos
token_regex =r'Capítulo \d+:'
```

```
#função para remocer os cabeçalhos
def quitar_cabecalho(texto):
    return re.sub(token_regex, '', texto).strip()
```

```
#aplicar re para cada capitulo
capitulo_regex =  [quitar_cabecalho(texto) for texto in corpus]
```

```
#Mostrar texto sem cabeçalho
for texto in capitulo_regex:
  print(texto)
```

#### 2) Compute a quantidade de caracteres, quantidade de palavras, quantidade de palavras únicas e quantidade de sentenças de cada capítulo e apresente em um `DataFrame` da biblioteca pandas

```
import re
import pandas as pd
```

```
len(texto)
```

```
#Função para contar carateres
def contarLetras(texto):
    return len(texto)
```

```
#Funcão para contar palavras
def contarPalavras(texto):
  palavras= re.findall(r'\b\w+\b',texto)
  return len(palavras)
```

```
#Função para contar palavras únicas
def contarPalavrasUnicas(texto):
    palavras = re.findall(r'\b\w+\b', texto)
    palavrasUnicas = set(palavras)
    return len(palavrasUnicas)
```

```
#Função para contar sentenças
def contarSentencas(texto):
    sentencas = re.split(r'[.!?]', texto)
    sentencas = [sentenca.strip() for sentenca in sentencas if sentenca.strip()]
    return len(sentencas)
```

```
dados = []
for i, texto in enumerate(corpus):
  caracteres = contarLetras(texto)
  palavras = contarPalavras(texto)
  palavrasUnicas = contarPalavrasUnicas(texto)
  sentencas = contarSentencas(texto)

```

```
  dados.append([f'Capitulo {i+1}',caracteres, palavras, palavrasUnicas, sentencas])
```

```
#Dataframe dos dados
df= pd.DataFrame(dados, columns=['Capitulo','Caracteres','Palavras','PalavrasUnicas','Sentencas'])
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy	as np
import pandas as pd
```

TF-IDF (Term Frequency-Inverse Document Frequency) é uma métrica utilizada para avaliar a importância de uma palavra em um documento dentro de um conjunto de documentos (corpus). O objetivo do TF-IDF é medir a relevância de uma palavra em um texto, levando em consideração a frequência dessa palavra em relação ao restante dos documentos.

```
# Função para calcular TF-IDF
def calcular_tfidf(corpus):
    vectorizer = TfidfVectorizer() #vectoriza com a função própria da lib
    X = vectorizer.fit_transform(corpus)
    df_tfidf = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
    return df_tfidf, vectorizer
```

TF-IDF Vectorizer: O TfidfVectorizer transforma o corpus de textos em uma matriz onde cada coluna é uma palavra e cada célula contém o valor da pontuação TF-IDF para aquela palavra em cada capítulo.
Cálculo do TF-IDF: O método fit_transform gera a matriz TF-IDF para o corpus s.
Busca da palavra: A função buscarPalavra recebe a palavra buscada, verifica se ela existe no corpus , e então procura o capítulo com a maior pontuação TF-IDF para essa palavra.
Ao final O capítulo com a maior pontuação e a pontuação correspondente são mostrados.
